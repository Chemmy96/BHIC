```python
import pandas as pd
import datetime
pd.set_option('display.max_columns', 500)
pd.set_option('display.max_rows', 200)
pd.set_option('display.width', 1000)
```

```python
data = pd.read_csv('data.csv', sep=',', encoding='latin-1')
dna = pd.read_csv('DNA.csv', sep=',', encoding='latin-1')
```

```python
data['year'] = data['date'].str.slice(5, 9).astype(float)
```

```python
dna_filtered = dna[dna['Beach'] == 'Bald Head Island'].sort_values(by='Year',ascending=True)
data_filtered = data[data['year'] >= 2010.0].sort_values(by='year')
dates = []
date_list = list(data_filtered['date'])
for i in date_list:
    dates.append(datetime.datetime.strptime(i, '%d%b%Y').strftime('%m/%d/%y'))

data_filtered['dates'] = dates
data_filtered['dates'] = data_filtered['dates'].str[1:]
data_filtered.head(1)
```

```python
dna_filtered.head(1)
```

# Create Unique ID

```python
data_filtered['yearstr'] = data_filtered['year']
data_filtered['yearstr'] = data_filtered['yearstr'].astype(int)
data_filtered['yearstr'] = data_filtered['yearstr'].astype(str)
data_filtered['neststr'] = data_filtered['Nest']
data_filtered['neststr'] = data_filtered['neststr'].astype(str)

data_filtered['id'] = data_filtered['yearstr'] + data_filtered['neststr']
data_filtered.head(1)
```

```python
dna_filtered['yearstr'] = dna_filtered['Year']
dna_filtered['yearstr'] = dna_filtered['yearstr'].astype(str)
dna_filtered['neststr'] = dna_filtered['Nest..']
dna_filtered['neststr'] = dna_filtered['neststr'].astype(str)

dna_filtered['id'] = dna_filtered['yearstr'] + dna_filtered['neststr']
dna_filtered
```

# Merge on ID

```python
merged_set = data_filtered.merge(dna_filtered, how='outer', on='id')
merged_set.head(2)
```
